```python

# Detecting the missing values of match
missing_values=["tie"]
data= pd.read_csv("IPL data.csv", na_values=missing_values)
data.isnull().sum()

```

```python

#Removing unwanted colomns in dataset
# Umpire1, umpire2, umpire3 comolomns elemenated.
data1= data.drop(['dl_applied','umpire1','umpire2','umpire3'],axis=1)
data1
```

```python
# Detecting the missing values of match
data1.isnull().sum()
```

```python
## Drop all null values
data2= data1.dropna()
data2
```

```python

###### Data Visualization ######

##seasonwise count of winner team
data3=data2.groupby('season')['winner'].value_counts()
data3
```

```python

year = 2008
data4 = pd.DataFrame(columns=['year', 'team', 'wins'])
for items in data3.iteritems():    
    if items[0][0]==year:
        print(items)
        win_series = pd.DataFrame({
            'year': [items[0][0]],
            'team': [items[0][1]],
            'wins': [items[1]]
        })
        data4 = data4.append(win_series)
        year += 1 
```

```python
data4
```

```python
mtp.title('Seasonwise which team win maximum number of matches')
sns.barplot('wins', 'team', hue='year', data=data4, palette='magma')
```

```python
# How the toss factor affects the match result
winning_count = 0
for index, value in data2.iterrows():
    if(value['toss_winner']==value['winner']):
#         print(value['winner'])
        winning_count+=1

print(f'The team who win the toss that win the match: {winning_count}')
prob = winning_count/len(data2)
print('How the toss factor affects the match result: {:.2f}' .format(prob))  
```
